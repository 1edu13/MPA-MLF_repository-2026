## Exercise 1.4 - The Fibonacci Sequence:

The Fibonacci Sequence is a series of numbers. The following number is found by adding up the two numbers before it. The first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13. The following number in this series above is 8 + 13 = 21


Your task is to implement a function *fibonacci* that takes an integer as an input and returns a list that contains all Fibonacci numbers with values lower than the input integer

Example:

*print(fibonaci(10))*

[0, 1, 1, 2, 3, 5, 8]


```python
def fibonacci(upper_threshold: int) -> list:
    # Handle the edge cases where the threshold is very small
    if upper_threshold <= 0:
        return []
    elif upper_threshold == 1:
        return [0]

    # Start our list with the first two Fibonacci numbers
    fib_list = [0, 1]

    # We use a while loop because we don't know exactly how many numbers we will generate
    while True:
        # Calculate the next number by adding the last two numbers in the list
        # We use negative indexing: -1 is the last item, -2 is the second to last
        next_fib = fib_list[-1] + fib_list[-2]

        # If the new number is equal to or greater than our threshold, we stop looping
        if next_fib >= upper_threshold:
            break

        # Otherwise, add the new number to the end of our list
        fib_list.append(next_fib)

    return fib_list

# Test the function exactly as requested in the instructions
print(fibonacci(10))
```

## Exercise 1.5. -  Rock, Paper, Scissors game:

#### basic

*Rock, Paper, Scissors* is a well-known and straightforward game. If you do not know the rules, google them. We will write a code for the Rock, Paper, Scissors game where the user plays against a random computer. The code can be written into one function *rock_paper_scissors*.

Notes:
- This implementation uses the **random** library to enable the computer to make a random choice.
- Keyword **Input** may be helpful. Check it
- After the function is run, the program will ask you about your movement, which you type in
- The game is case-insensitive for user input (e.g., "rock", "Rock", and "ROCK" are all valid).
- This script plays one round of the game.
- The script prints the result ('You lose', 'You win', 'It is a tie')


```python
import random

def rock_paper_scissors() -> None:
    # Define the valid options for the game
    options = ['rock', 'paper', 'scissors']

    # The computer makes a random choice from the list
    computer_choice = random.choice(options)

    # Get the user's input and immediately convert it to lowercase
    # This handles "Rock", "ROCK", "rock", etc.
    user_input = input("Enter your move (rock, paper, scissors): ").strip().lower()

    # Check if the user entered a valid option
    if user_input not in options:
        print("Invalid input. Please try again.")
        return

    print(f"Computer chose: {computer_choice}")

    # Determine the winner
    if user_input == computer_choice:
        print("It is a tie")
    elif (user_input == 'rock' and computer_choice == 'scissors') or \
         (user_input == 'paper' and computer_choice == 'rock') or \
         (user_input == 'scissors' and computer_choice == 'paper'):
        print("You win")
    else:
        # If it's not a tie and you didn't win, you must have lost!
        print("You lose")

# Run the game
rock_paper_scissors()
```

#### Advanced

Extend the Rock, Paper, Scissors game to be able to play *n* rounds.

Notes:
- The scrips will contain two counters - user score and computer score
- After each run, the script prints the current scores and rounds.
- When the number of rounds is reached, the script prints the result ('You lose', 'You win', 'It is a tie')

```python
import random

def rock_paper_scissors_advanced(n_rounds: int) -> None:
    options = ['rock', 'paper', 'scissors']

    # Initialize our score counters
    user_score = 0
    computer_score = 0

    # Loop for the number of rounds requested
    for current_round in range(1, n_rounds + 1):
        print(f"\n--- Round {current_round} of {n_rounds} ---")

        # We use a while loop here to make sure the user enters a valid option
        # so they don't accidentally waste a round on a typo!
        user_input = ""
        while user_input not in options:
            user_input = input("Enter your move (rock, paper, scissors): ").strip().lower()
            if user_input not in options:
                print("Invalid input. Please try again.")

        computer_choice = random.choice(options)
        print(f"Computer chose: {computer_choice}")

        # Determine the winner of the round and update scores
        if user_input == computer_choice:
            print("It is a tie for this round!")
        elif (user_input == 'rock' and computer_choice == 'scissors') or \
             (user_input == 'paper' and computer_choice == 'rock') or \
             (user_input == 'scissors' and computer_choice == 'paper'):
            print("You win this round!")
            user_score += 1
        else:
            print("You lose this round!")
            computer_score += 1

        # Print the current scores and rounds as requested
        print(f"Current Score -> You: {user_score} | Computer: {computer_score}")

    # After all rounds are finished, print the final overall result
    print("\n=== FINAL RESULT ===")
    if user_score > computer_score:
        print("You win")
    elif user_score < computer_score:
        print("You lose")
    else:
        print("It is a tie")

# Test the function for a "best of 3" match
rock_paper_scissors_advanced(3)
```

## Exercise 2.1


In this exercise, you will work with NumPy arrays and learn how to efficiently modify and process them using both loops and vectorized operations. Your task is to:

1. Implement the function *create_array_nxn* that generates an n×n NumPy array filled with numbers from n² - 1 down to 0
2. Using a loop-based approach (for-loop, while-loop), implement the function *apply_threshold_loop* that takes any NumPy array and replaces all numbers below a user-defined threshold with 0.
3. Implement function *apply_threshold_vectorized* that performs the same operation as *apply_threshold_loop*, but without loops, using NumPy vectorized operations
4. Implement function *compare_performance* that compares the execution time of functions *apply_threshold_vectorized* and *apply_threshold_loop* with **time** library and print the results


**Additional Notes**:

- Ensure n is a positive integer (n > 0).
- Use the time library to measure execution times for performance comparison.
- The function *compare_performance(n, threshold)* should:
    - Generate an n×n array.
    - Apply both loop-based and vectorized thresholding.
    - Print execution times.

```python
import numpy as np
import time
```

```python
# 1. Generate an n x n array
def create_array_nxn(n: int) -> np.ndarray:
    # Calculate the maximum number (n squared minus 1)
    max_num = (n**2) - 1

    # np.arange creates a 1D array from max_num down to 0 (step is -1)
    # Then we use .reshape(n, n) to turn it into a 2D grid
    arr = np.arange(max_num, -1, -1)
    return arr.reshape(n, n)
```

```python
# 2. Apply threshold using a standard nested for-loop
def apply_threshold_loop(arr: np.ndarray, threshold: int) -> np.ndarray:
    # We make a copy so we don't accidentally modify the original array
    result = arr.copy()
    rows, cols = result.shape

    # Loop through every single row and column one by one
    for i in range(rows):
        for j in range(cols):
            if result[i, j] < threshold:
                result[i, j] = 0

    return result
```

```python
# 3. Apply threshold using NumPy's vectorized operations
def apply_threshold_vectorized(arr: np.ndarray, threshold: int) -> np.ndarray:
    result = arr.copy()

    # This is the "vectorized" way! No loops needed.
    # It finds all positions where the condition is True and sets them to 0 instantly.
    result[result < threshold] = 0

    return result
```

```python
# 4. Compare the performance of both methods
def compare_performance(n: int, threshold: int) -> None:
    # Create the test array
    test_array = create_array_nxn(n)

    # Test the loop method
    start_time_loop = time.time()
    apply_threshold_loop(test_array, threshold)
    end_time_loop = time.time()
    loop_duration = end_time_loop - start_time_loop

    # Test the vectorized method
    start_time_vec = time.time()
    apply_threshold_vectorized(test_array, threshold)
    end_time_vec = time.time()
    vec_duration = end_time_vec - start_time_vec

    # Print the results
    print(f"Testing with a {n}x{n} array (Total elements: {n**2})")
    print(f"Loop execution time:       {loop_duration:.6f} seconds")
    print(f"Vectorized execution time: {vec_duration:.6f} seconds")

# Let's run the comparison! Try a relatively large number like 1000 to see the difference clearly.
compare_performance(1000, 500000)
```

```python
def show_in_digi(input_integer: int) -> None:
    str_input = str(input_integer)
    arrays_to_concat = []

    # Create a spacer (5 rows, 1 column of zeros) for the gap between digits
    spacer = np.zeros((5, 1))

    for index, digit in enumerate(str_input):
        digit_array = numbs[digit]
        arrays_to_concat.append(digit_array)

        if index < len(str_input) - 1:
            arrays_to_concat.append(spacer)

    final_image = np.concatenate(arrays_to_concat, axis=1)

    # cmap='binary' gives us the black/white look
    # interpolation='nearest' keeps the blocks perfectly crisp
    plt.imshow(final_image, cmap='binary', interpolation='nearest')

    # Hide the axes so it looks like a clean image, not a graph
    plt.axis('on')

    plt.show()

# Run it again to see the crisp, black-and-white output!
show_in_digi(5289)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# --- 0. Load the Dataset ---
# Colab has this dataset pre-installed in the 'sample_data' directory
file_path = "sample_data/california_housing_train.csv"
dataset = pd.read_csv(file_path)

print("--- 1. What dataset.describe() does ---")
# dataset.describe() calculates summary statistics for numerical columns
# (like count, mean, min, max, standard deviation, and percentiles).
# It's a great way to quickly understand the distribution of your data!
display(dataset.describe())

print("\n--- 2. Rows where total_bedrooms > 310 ---")
# We use boolean indexing to filter the dataset
high_bedrooms = dataset[dataset['total_bedrooms'] > 310]
display(high_bedrooms.head()) # Using .head() just to keep the output short

print("\n--- 3. Drop the first and last row ---")
# .iloc stands for integer-location based indexing.
# [1:-1] means "start at index 1 (the second row) and stop at -1 (the last row, exclusive)"
dataset = dataset.iloc[1:-1]
print(f"New dataset shape: {dataset.shape}")

print("\n--- 4. Plot Households and their Mean ---")
# Calculate the mean
mean_households = dataset['households'].mean()

plt.figure(figsize=(10, 5))
# Plot the dots (x-axis = index/ID, y-axis = households)
plt.scatter(dataset.index, dataset['households'], color='blue', alpha=0.3, label='Households')
# Draw a horizontal line for the mean
plt.axhline(mean_households, color='red', linewidth=2, label=f'Mean: {mean_households:.2f}')

plt.title("Number of Households per Record")
plt.xlabel("ID (Index)")
plt.ylabel("Households")
plt.legend()
plt.show()

print("\n--- 5. Check and replace NaN (missing) values ---")
# .isna().sum() shows how many missing values are in each column
missing_values = dataset.isna().sum()
print("Missing values before:\n", missing_values)

# Replace NaNs with the mean of each respective column
# (Note: This Colab sample dataset usually has no NaNs, but this is how you'd do it!)
dataset = dataset.fillna(dataset.mean())

print("\n--- 6. Plot Latitude vs Longitude ---")
# This actually plots a rough map of California!
plt.figure(figsize=(8, 6))
# The columns in the dataset are typically named 'latitude' and 'longitude'
plt.scatter(dataset['longitude'], dataset['latitude'], alpha=0.2, color='green')
plt.title("Map of California Housing (Long vs Lat)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()

print("\n--- 7. Min-Max Normalization ---")
# Min-Max Normalization scales values so they fit exactly between 0 and 1.
# Formula: (value - min) / (max - min)
col1, col2 = 'median_income', 'median_house_value'

# Normalize column 1
dataset[col1 + '_normalized'] = (dataset[col1] - dataset[col1].min()) / (dataset[col1].max() - dataset[col1].min())

# Normalize column 2
dataset[col2 + '_normalized'] = (dataset[col2] - dataset[col2].min()) / (dataset[col2].max() - dataset[col2].min())

print(dataset[[col1, col1 + '_normalized', col2, col2 + '_normalized']].head())

print("\n--- 8. Correlation Matrix ---")
# A correlation matrix shows how linearly related every column is to every other column.
# Values are between -1 (perfect negative correlation) and 1 (perfect positive correlation).
corr_matrix = dataset.corr()
display(corr_matrix)
```
